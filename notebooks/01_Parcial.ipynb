```python
#importamos librerias 
import pandas as pd 
import numpy as np 
from pathlib import Path 
```

```python
#Links a los conjuntos de datos
url_cgsse = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/cgsse.CSV'
url_ocupados = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/Ocupados.CSV'
url_ft = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/Fuerza%20de%20trabajo.CSV'
url_desocupados = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/No%20ocupados.CSV'

# CGSSE
df_cgsse = pd.read_csv(url_cgsse, encoding='latin-1', sep=';', 
                       low_memory=False)

#Ocupados
df_ocupados=pd.read_csv(url_ocupados,encoding='latin-1',sep=';',
                        low_memory=False)

#Fuerza de trabajo
df_ft=pd.read_csv(url_ft,
                  encoding='latin-1',sep=';',low_memory=False)
#No ocupados
df_desocupados=pd.read_csv(url_desocupados,encoding='latin-1',
                           sep=';',low_memory=False)
```

```python
#1. Población total
PT =  df_cgsse['FEX_C18'].sum()
print(f' R/ La población total es de {PT} millones de personas')

#Población en edad de trabajar
PET = df_cgsse[df_cgsse['P6040']>=15]['FEX_C18'].sum()
print(f' R/ La población en edad de trabajar es de {PET} millones de personas')

#Población que no está en edad de trabajar
PNET = df_cgsse[df_cgsse['P6040']<=15]['FEX_C18'].sum()
print(f' R/ La población que no está en edad de trabajar es de {PNET} millones de personas')

#Población económicamente activa
PEA=df_ft[(df_ft['FT']==1)]['FEX_C18'].sum()
print(f' R/ Población Economicamente Activa {PEA} millones de personas')

#FFT Pertenece a la PEI  si:1
PEI=df_ft[(df_ft['FFT']==1)]['FEX_C18'].sum()
print(f' R/ Población Economicamente Inactiva {PEI}')

#OCI : Ocupado ? 1:si
O=df_ocupados[(df_ocupados['OCI']==1)]['FEX_C18'].sum()
TO=O/PET*100
print(f' R/ Tasa de ocupacion para Diciembre de 2024 {round(TO,1)}')

#OCI : Ocupado ? 1:si
D=df_desocupados[(df_desocupados['DSI']==1)]['FEX_C18'].sum()
TD=D/PEA*100
print(f' R/ Tasa de desempleo para diciembre de 2024 {round(TD,1)}')
```

```python
# 8. ¿Cuántas personas hay en el Quindío en condición de desempleo entre los 18 y los 30 años?


#Población en el Quindío entre 18 y 30 que estan desempleados

PQD = df_desocupados[(df_desocupados['DSI']==1) & (df_desocupados['DPTO']==63) & (df_desocupados['P6040']>=18) & (df_desocupados['P6040']<=30)]['FEX_C18'].sum()
 
print (f' población economicamente activa en el Quindío entre 18 y 30 años desempleados: {PQD}')

PEI = df_ft[(df_ft['FFT']==1) & (df_ft['DPTO']==63) & (df_ft['P6040']>=18) & (df_ft['P6040']<=30)]['FEX_C18'].sum()

print(f'Poblacion Economicamente Inactiva entre 18 y 30 años en Quindío: {int(PEI)}')

PTDQ = PQD + PEI
print(f'Poblacion desempleada entre 18 y 30 años en Quindío: {PTDQ}')
```

```python
# ¿Cuántas personas hay ocupadas en el departamento del Tolima , 
# cuyas edades están entre los 20 y los 25 años , qué porcentaje de 
# la población total del Tolima representan ? 

PFTt = df_ft[(df_ft['FT']==1) & (df_ft['DPTO']==73) & (df_ft['P6040']>=20) & (df_ft['P6040']<=25)]['FEX_C18'].sum()
print (f' población del departamento del Tolima economicamente activa en edad entre los 20 y 25 años que hacen parte de la fuerza de trabajo: {PFTt}')

PDt = df_desocupados[(df_desocupados['DSI']==1) & (df_desocupados['DPTO']==73) & (df_desocupados['P6040']>=20) & (df_desocupados['P6040']<=25)]['FEX_C18'].sum()
print (f' población en el Tolima entre 20 y 25 años desocupados: {PDt}')

# Ft = O + D -> O = FT - D

OT = PFTt - PDt
print (f' población en el Tolima entre 20 y 25 años ocupados: {OT}')

# Población total del Tolima

# Población del departamento del Tolima
PTt = df_cgsse[df_cgsse['DPTO']==73]['FEX_C18'].sum()
print(f' Población del departamento del Tolima: {PTt}')

# Porcentaje población ocupada respecto al total de población

PPOt = OT/PTt
print(f' Representación porcentual de población entre 20 y 25 años ocupada en el Tolima: {PPOt}%')
```
